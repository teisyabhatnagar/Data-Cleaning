```python
foods=pd.read_csv('/Users/teisya/Desktop/Python 2 Datasets/restaurant_foods.csv')
customer=pd.read_csv('/Users/teisya/Desktop/Python 2 Datasets/restaurant_customers.csv')
week1=pd.read_csv('/Users/teisya/Desktop/Python 2 Datasets/restaurant_week_1_sales.csv')
week2=pd.read_csv('/Users/teisya/Desktop/Python 2 Datasets/restaurant_week_2_sales.csv')
```

```python
week1.head()
```

```python
foods.head(5)
```

```python
#this is left merge
```

```python
week1.merge(foods, how="left", on="Food ID") #because its merging with left table 
```

```python
week2.head()
```

```python
customer.head()
```

```python
#merge customer ID and ID and merge on right side
```

```python
week2.merge(customer, how="left", left_on="Customer ID", right_on="ID") #if both columns are different, both cust  id and id there
```

```python
week2.merge(customer, how="left", left_on="Customer ID", right_on="ID").drop("ID", axis="columns")
```
